**Q3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title. <br>
Test Data: <br>
test.txt  <br>
1 2 <br>
2 4 <br>
3 1**

```python
def read_data_from_file(file_path):
    x = []
    y = []
    with open(file_path, 'r') as file:
        for line in file:
            values = line.split()
            x.append(float(values[0]))
            y.append(float(values[1]))
    return x, y

file_path = 'test.txt'

data = """1 2
2 4
3 1"""

with open(file_path, 'w') as file:
    file.write(data)

x, y = read_data_from_file(file_path)

plt.plot(x, y, marker='o',c='black',ls='--',lw='5',markerfacecolor='blue', markersize=12)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot from File Data')
plt.show()
```

**Q12. Write a Python program to create multiple types of charts**

```python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 3, 5, 7, 11])
y2 = np.array([1, 4, 6, 8, 10])
categories = ['A', 'B', 'C', 'D']
values = [23, 17, 35, 29]

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y1, marker='o', label='Line Chart')
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2, 2, 2)
plt.bar(categories, values, color='skyblue', label='Bar Chart')
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()

plt.subplot(2, 2, 3)
plt.scatter(x, y2, color='green', label='Scatter Plot')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2, 2, 4)
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Pie Chart')

plt.tight_layout()
plt.show()
```

**Q13. Write a Python program to display the grid and draw line charts of the closing value of <br>

**Q14. Write a Python program to display the grid and draw line charts of the closing value of <br>
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with <br>
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn <br>
off ticks.**

```python
import matplotlib.dates as mdates
from datetime import datetime

dates = [
    datetime.strptime('03-10-16', '%d-%m-%y'),
    datetime.strptime('04-10-16', '%d-%m-%y'),
    datetime.strptime('05-10-16', '%d-%m-%y'),
    datetime.strptime('06-10-16', '%d-%m-%y'),
    datetime.strptime('07-10-16', '%d-%m-%y')
]
closes = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

plt.figure(figsize=(10, 6))
plt.plot(dates, closes, marker='o', linestyle='-', color='r', label='Closing Value')

plt.grid(True, which='major', linestyle='-', linewidth=0.7, color='blue')
plt.grid(True, which='minor', linestyle=':', linewidth=0.5, color='gray')

plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False)

plt.title('Alphabet Inc. Closing Values (Oct 3, 2016 - Oct 7, 2016)')
plt.xlabel('Date')
plt.ylabel('Closing Value')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%y'))

plt.legend()
plt.show()
```

**Q15. Write a Python program to create multiple plots.**

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1) 
plt.plot(x, y1, label='Sine', color='blue')
plt.title('Sine Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2, 2, 2)  
plt.plot(x, y2, label='Cosine', color='green')
plt.title('Cosine Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2, 2, 3) 
plt.plot(x, y3, label='Tangent', color='red')
plt.title('Tangent Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()


plt.subplot(2, 2, 4) 
plt.scatter(x, y1, label='Scatter Sine', color='purple')
plt.title('Scatter Plot of Sine')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()


plt.tight_layout()
plt.show()
```
