**Q1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.**

```python
import matplotlib.pyplot as plt
import random

x = [random.uniform(0, 1) for _ in range(50)]
y = [random.uniform(0, 1) for _ in range(50)]

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', edgecolors='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Random Distribution')
plt.show()
```

**Q2. Write a Python program to draw a scatter plot with empty circles taking a random distribution
in X and Y and plotted against each other.**

```python
x = [random.uniform(0, 1) for _ in range(50)]
y = [random.uniform(0, 1) for _ in range(50)]

plt.figure(figsize=(8, 6))
plt.scatter(x, y, facecolors='none', edgecolors='blue', alpha=0.7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Random Distribution with Empty Circles')
plt.show()
```

**Q3. Write a Python program to draw a scatter plot using random distributions to generate balls of
different sizes.**

```python
x = [random.uniform(0, 1) for _ in range(50)]
y = [random.uniform(0, 1) for _ in range(50)]
sizes = [random.uniform(20, 200) for _ in range(50)]

plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, color='blue', edgecolors='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Random Ball Sizes')
plt.show()
```

**Q5. Write a Python program to draw a scatter plot for three different groups camparing weights
and heights.**

```python
group1_weights = [random.uniform(50, 80) for _ in range(10)]
group1_heights = [random.uniform(150, 180) for _ in range(10)]
group2_weights = [random.uniform(60, 90) for _ in range(10)]
group2_heights = [random.uniform(160, 190) for _ in range(10)]
group3_weights = [random.uniform(70, 100) for _ in range(10)]
group3_heights = [random.uniform(170, 200) for _ in range(10)]

plt.figure(figsize=(8, 6))
plt.scatter(group1_weights, group1_heights, color='blue', edgecolors='black', label='Group 1', alpha=0.7)
plt.scatter(group2_weights, group2_heights, color='red', edgecolors='black', label='Group 2', alpha=0.7)
plt.scatter(group3_weights, group3_heights, color='green', edgecolors='black', label='Group 3', alpha=0.7)

plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('Scatter Plot of Weights vs Heights for Three Groups')
plt.legend()
plt.show()
```
